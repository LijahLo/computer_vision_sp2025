```python
import sys
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_image(image1, strng):
    image = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB) ##BGR to RGB conversion for display using Matplotlib
    plt.title(strng)
    plt.imshow(image)
    plt.show()
```

```python
img = cv2.imread("Entlebucher.jpg")
print("height, width, channels: ", img.shape)
print("total pixels: ", img.size)
print("Pixel intensity at y=110, x =100: ", img[110,100])
display_image(img, 'Original')
```

```python
## red channel
b,g,r = cv2.split(img)
red = np.zeros_like(img)
red[:, :, 2] = r

display_image(red,'only red channel')
```

```python
img_red = img.copy()
img_red[:,:,0] =0
img_red[:,:,1] =0


display_image(img_red, "zeroing out blue and green")
```

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display_image(imgGray, "Grayscale")
```

```python
img_roi = img[80:100, 80:95]
display_image(img_roi, 'Experiments with image ROIs')
```
