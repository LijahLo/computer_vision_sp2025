```python
import sys
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Entlebucher.jpg") 
print ("Gives the shape of dog image (height,width,channels):",img.shape)
print ("Get a particular pixel value at location x = 100, y = 110", img[100,110])
#Reading Entlebucher.jpg image on OpenCV
```

```python
##Function to display image using Matplotlib 
def display_image(image1,strng):
    image = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)##BGR to RGB conversion for displaying using Matplotlib 
    plt.title(strng) 
    plt.imshow(image)
    plt.show() 
print ("Shape & Size & Type: ", img.shape, img.size, img.dtype)
display_image(img,'Original image') #original image 
```

Answer to Question 3: The height of the dog image is 480. The width of the dog image is 555. The number of channels is 3. The total number of pixels in the dog image is 799200 units. The pixel itensity at position x = 100, y = 110 is [61 40 32]

```python
b,g,r = cv2.split(img) # using OpenCV to split channels
img_red = np.zeros_like(img) # initializing an image with same dimensions as original image
img_red[:,:, 0] = r # note that we are using RGB convention and not OpenCV here. 

plt.imshow(img_red)
plt. show()
```

Answer to Question 4: What I see is the original image but it is shaded with a red overlay on top of the dog image. The reason why is because I filtered out the shades of green and blue within the RGB color set. 


```python
imgCopy = img.copy() # make a copy of the image for processing.
imgCopy[:,:, 1] = 0 ## making all pixels in the green channel 0, meaning we are removing the green channel.
imgCopy[:,:,0] = 0 ## making all pixels in the blue channel 0, meaning we are removing the blue channel.
display_image(imgCopy, 'Green and Blue channel altered') ## displaying the image after zeroing the green and blue channel 
```

Answer to Question 6: The difference between question 4 and 5 is that these are 2 different methods of code to get the same results of the red channel overlay the dog image. 

```python
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print ("Shape & Size & Type: ", imgGray.shape, imgGray.size, imgGray.dtype)
display_image(imgGray,'Grayscale image')
```

```python
img_roi = img[70:100, 60:130]
display_image(img_roi, 'Experiments with image ROIs')
```
